```python
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='lead_time', bins=50, kde=True)
plt.title('Lead Time Distribution')
plt.xlabel('Lead Time (days)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='is_canceled', y='lead_time')
plt.title('Cancellation Rate by Lead Time')
plt.xlabel('Cancellation (0: Not Canceled, 1: Canceled)')
plt.ylabel('Lead Time (days)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

filtered_data = data.dropna(subset=['adults', 'children', 'babies'])

# Create subplots for each variable
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Plot distribution of adults
sns.histplot(filtered_data['adults'], bins=10, kde=True, ax=axs[0])
axs[0].set_title('Distribution of Adults')

# Plot distribution of children
sns.histplot(filtered_data['children'], bins=10, kde=True, ax=axs[1])
axs[1].set_title('Distribution of Children')

# Plot distribution of babies
sns.histplot(filtered_data['babies'], bins=5, kde=True, ax=axs[2])
axs[2].set_title('Distribution of Babies')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='country', order=data['country'].value_counts().iloc[:10].index)
plt.title('Top 10 Countries of Guests')
plt.xlabel('Country')
plt.ylabel('Number of Guests')
plt.xticks(rotation=45)
plt.show()
```

```python
top_10_countries = data['country'].value_counts().head(10).index

plt.figure(figsize=(12, 6))
sns.barplot(data=data[data['country'].isin(top_10_countries)], 
            x='country', 
            y='is_canceled', 
            estimator=lambda x: sum(x) / len(x) * 100)
plt.title('Cancellation Rates by Country (Top 10)')
plt.xlabel('Country')
plt.ylabel('Cancellation Rate (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='reservation_status')
plt.title('Distribution of Reservation Statuses')
plt.xlabel('Reservation Status')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='reservation_status', hue='is_canceled')
plt.title('Reservation Status vs. Cancellation')
plt.xlabel('Reservation Status')
plt.ylabel('Count')
plt.legend(title='Cancellation', loc='upper right')
plt.xticks(rotation=45)
plt.show()
```

```python
months_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
data['arrival_date_month'] = pd.Categorical(data['arrival_date_month'], categories=months_order, ordered=True)

# Create a bar plot of bookings per month
plt.figure(figsize=(12, 6))
sns.countplot(x='arrival_date_month', data=data, hue='is_canceled', palette='viridis')
plt.title('Bookings per Month')
plt.xlabel('Month')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x = 'hotel', hue = 'is_canceled' , data = data, palette = "BuPu")
labels = ['not_cancelled','cancelled' ]
plt.legend(labels)
```

```python
cancelled_reservations = data[data['is_canceled'] == 1]
not_cancelled_reservations = data[data['is_canceled'] == 0]

top_10_countries = (cancelled_reservations['country'].value_counts()[:10])
plt.figure(figsize = (20,8))
plt.title("Top 10 country hotels with cancelled reservations")
plt.pie(top_10_countries,labels = top_10_countries.index, autopct = '%.2f',
)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import math

numerical_columns = [col for col in data.columns if data[col].dtype in ['int64', 'float64'] and col != 'id']

num_cols = len(numerical_columns)
num_rows = math.ceil(num_cols / 3) 

plt.figure(figsize=(15, 5*num_rows)) 
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(num_rows, 3, i) 
    sns.boxplot(data[column])
    plt.title(column)
plt.tight_layout()
plt.show()
```
